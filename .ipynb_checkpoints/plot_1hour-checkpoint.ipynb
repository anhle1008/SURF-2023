```python
import numpy as np
import matplotlib.pyplot as plt

def read_data(filename):
    with open(filename, 'r') as file:
        next(file)
        contents = file.readlines()
        
    output = []
    for line in contents:
        m, o, r = line.strip().split()
        output.append([float(m), float(o), float(r)])
    
    return np.array(output).T
```

```python
time, offset, rtt = read_data('data_1_hour.txt')
start = time[0]
time -= start
plt.plot(time, offset, label='time offset')
plt.plot(time, rtt, label='round-trip delay')
plt.title('Offset and Delay between NTP and my laptop in 1 hour')
plt.gca().get_yaxis().get_major_formatter().set_useOffset(False)
plt.xlabel('Hours')
plt.xlim([0, max(time)])
plt.ylabel('Seconds')
plt.grid(True)
plt.legend()
plt.show()
```

```python
"""Ignore 10 outliers"""
diff = list(offset)
delay = list(rtt)
num = 10
for i in range(num):
    m1 = diff.index(min(diff))
    diff[m1] = diff[0]
    m2 = delay.index(max(delay))
    delay[m2] = delay[0]
    
plt.plot(time, diff, label='time offset')
plt.plot(time, delay, label='round-trip delay')
plt.title('Offset and Delay between NTP and my laptop in 1 hour')
plt.gca().get_yaxis().get_major_formatter().set_useOffset(False)
plt.xlabel('Hours')
plt.xlim([0, max(time)])
plt.ylabel('Seconds')
plt.grid(True)
plt.legend()
plt.show()
```

```python
coeffs1 = np.polyfit(time, diff, deg=2)
model1 = np.polyval(coeffs1, time)
resid1 = diff - model1

coeffs2 = np.polyfit(time, delay, deg=2)
model2 = np.polyval(coeffs2, time)
resid2 = delay - model2

fig, ax = plt.subplots(1,2,figsize=(12,4))

ax[0].plot(time, resid1)
ax[0].set_xlabel('Hours')
ax[0].set_ylabel('Seconds')
ax[0].set_title('Time offset residuals')
ax[0].grid('on', 'both')

ax[1].plot(time, resid2)
ax[1].set_xlabel('Hours')
ax[1].set_ylabel('Seconds')
ax[1].set_title('Round-trip delay residuals')
ax[1].grid('on', 'both')

plt.subplots_adjust(wspace=0.3)
plt.show()
```

```python
import allantools

# Calculate residuals
resid1 = diff - model1
resid2 = delay - model2

taus1, allan_var1, _, _ = allantools.oadev(resid1, rate=1/60, data_type='phase') # 1-min intervals 
taus2, allan_var2, _, _ = allantools.oadev(resid2, rate=1/60, data_type='phase') # 1 sec = 1/60 minutes

fig, ax = plt.subplots(1,2,figsize=(12,4))

# Second column data
ax[0].loglog(taus1, allan_var1, 'o')
ax[0].set_xlabel('Averaging Time , Seconds')
ax[0].set_ylabel('Allan deviation ()')
ax[0].set_title('Frequency Stability of Time Offset')
ax[0].set_ylim([1e-6, 1e-3])
ax[0].grid('on', 'both')
# Time column data
ax[1].loglog(taus2, allan_var2, 'o')
ax[1].set_xlabel('Averaging Time , Seconds')
ax[1].set_ylabel('Allan deviation ()')
ax[1].set_ylim([1e-6, 1e-3])
ax[1].set_title('Frequency Stability of RRT')
ax[1].grid('on', 'both')

plt.subplots_adjust(wspace=0.3)
plt.show()
```
